### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total_scores'] = df.math_score+df.reading_score+df.writing_score
df['avg_scores']= df.total_scores/3
df.head(1)
```

```python
# check the number of students with outof performance
math_full= df[df.math_score ==100]['total_scores'].count()
writing_full= df[df.writing_score ==100]['total_scores'].count()
reading_full= df[df.reading_score ==100]['total_scores'].count()
print(f"Number of studenyts with full marks in math: {math_full}")
print(f"Number of studenyts with full marks in writing: {writing_full}")
print(f"Number of studenyts with full marks in reading: {reading_full}")
```

```python
# check the number of students with poor performance
poor_performance_marks=20
math_poor= df[df.math_score <= poor_performance_marks]['total_scores'].count()
writing_poor= df[df.writing_score <= poor_performance_marks]['total_scores'].count()
reading_poor= df[df.reading_score <= poor_performance_marks]['total_scores'].count()
print(f"Number of studenyts with poor marks in math: {math_full}")
print(f"Number of studenyts with poot marks in writing: {writing_poor}")
print(f"Number of studenyts with poor marks in reading: {reading_poor}")
```

#### insights:
- from the above values students performed the best in reading 
- poor performance in math

### 4.4 Feature Wise Visualization

#### 4.4.1 GENDER COLUMN
- How is distribution of Gender ?
- Is gender has any impact on student's performance ?
#### UNIVARIATE ANALYSIS ( How is distribution of Gender ? )

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

#### Insights

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? )

```python
gender_group = df.groupby(['gender']).agg(
    Total_Average=('total_scores','mean'),
    Math_Average=('math_score','mean'),
    Writing_Average=('writing_score','mean'),
    Reading_Average=('reading_score','mean')	
    
)
gender_group
```

```python



plt.figure(figsize=(10, 8))

X = ['Total_Average','Math_Average','Writing_Average','Reading_Average']


female_scores = [gender_group['Total_Average'][0], gender_group['Math_Average'][0],gender_group['Writing_Average'][0],gender_group['Reading_Average'][0]]
male_scores = [gender_group['Total_Average'][1], gender_group['Math_Average'][1],gender_group['Writing_Average'][1],gender_group['Reading_Average'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights
- on an average females have a better overall score than men.
- whereas males have scored higher in Maths.
